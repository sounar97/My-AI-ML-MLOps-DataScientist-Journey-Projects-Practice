```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
ages=[random.randrange(25,35,1) for ages in range(11)]
ages=sorted(ages,reverse=False)
```

```python
devs=[random.randrange(30000,45000,1) for devs in range(11)]
devs=sorted(devs,reverse=False)
```

```python
plt.plot(ages,devs)
plt.show
```

```python
plt.plot(ages,devs)
plt.title("Random number")
plt.xlabel("Ages")
plt.ylabel("Devs")
plt.show()
```

```python
py_devs=[random.randrange(50000,100000,1) for devs in range(11)]
py_devs=sorted(devs,reverse=False)
```

```python
plt.plot(ages,devs,color="#FF33E9",linestyle="--")
plt.plot(ages,py_devs,color="#3344FF",linestyle="--")
plt.title("Median Salary in $ with Age")
plt.xlabel("Age")
plt.ylabel("Median Salary")
plt.plot(devs,py_devs)
plt.show()
```

```python
plt.plot(devs,py_devs,linewidth=3)
plt.show()
```

```python
#Adding grid to the plot
plt.plot(devs,py_devs,linewidth=3)
plt.grid(True)
plt.show()
```
