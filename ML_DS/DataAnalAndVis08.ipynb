```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
breast_data = breast.data
print(breast_data)
print(breast_data.shape)
```

```python
breast_lables = breast.target
print(breast_lables)
print(breast_lables.shape)
```

```python
#Reshape the dataset by adding label to it.Concatenate the data with the label
labels = np.reshape(breast_lables,(569,1))
final_breast_data = np.concatenate([breast_data,labels],axis=1)
print(final_breast_data.shape)
```

```python
#Create a dataframe
breast_dataset = pd.DataFrame(final_breast_data)
print(breast_dataset.head())
```

```python
#Print the features that are in the dataset
features = breast.feature_names
print(features)
```

```python
features_labels = np.append(features,'labels')
```

```python
#Embedding the column names into the datframe
breast_dataset.columns = features_labels
breast_dataset.head()
```

```python
breast_dataset['labels'].replace(0, 'Benign',inplace=True)
breast_dataset['labels'].replace(1, 'Benign',inplace=True)
breast_dataset.tail()
```

```python
#You start by the standerdizin gthe data
from sklearn.preprocessing import StandardScaler
```

```python
#You start by the standerdizin gthe data
from sklearn.preprocessing import StandardScaler
x = breast_dataset.loc[:,features].values
x=StandardScaler().fit_transform(x)
print(x.shape)
```

```python
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
```

```python
normalized_breast =pd.DataFrame(x,columns=feat_cols)
print(normalized_breast)
```

```python
normalized_breast.tail()
```

```python
#Projecting the thirty dimensional Breast Cancer Data to two dimensional principle components
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_DF = pd.DataFrame(data = principalComponents_breast , columns = ['principal Component 1','principal component 2'])
principal_breast_DF.tail()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(index,principal Component 1,marker='D',color='cyan',linestle='.-',linewidth=2)
plt.plot(index,principal component 2,marker='*',color='green',linestle='--',linewidth=2)
plt.scatter(principal_breast_DF)
plt.xlabel('principal component 1')
plt.ylabel('principal component 2')
plt.show()
```
