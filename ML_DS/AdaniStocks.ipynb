```python

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
# Select features for segmentation
features = ["open", "high", "low", "volume", "dividends", "stock_splits"]

# Choose the number of clusters (k) based on your analysis or domain knowledge
k = 3  # adjust this value

# Initialize and fit the K-Means model
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data[features])

# Assign cluster labels to each data point
data["cluster"] = kmeans.labels_

# Print basic cluster information
print(f"Number of clusters: {k}")
for i in range(k):
    cluster_data = data[data["cluster"] == i]
    print(f"\nCluster {i+1}:")
    print(f"- Average open price: {cluster_data['open'].mean():.2f}")
    print(f"- Average high price: {cluster_data['high'].mean():.2f}")
    print(f"- Average volume: {cluster_data['volume'].mean():.2f}")
```

```python
sumTM=1000000#total amount invested

s1=100#number of shares owned
for i in range(len(df)):
  if df.loc[i,'Day']==30:
    sumTM+=df.loc[i,'open']
    s1+=1

print("Total Invested= Rs",round(sumTM,2))
print("Shares Owned of Adani =",s1)
print("Average Investmentment of 1 share = Rs",round((sumTM/s1),2))
tm_end=227
result1=round((tm_end*s1)-sumTM,2)
roiTM=round((result1/sumTM)*100,2)
print("Investment Result:")
if result1<0:

    print("Net Unrealised Loss = Rs",result1)

else:

    print("Net Unrealised Profit = Rs",result1)



print(" ROI =",roiTM,"%")
```
